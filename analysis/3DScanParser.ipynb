```python
from scipy.spatial.transform import Rotation
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def getDistance(sensorInput):
    return sensorInput
```

```python
data = np.genfromtxt('logFile.csv', delimiter=',')
theta1=data[:,0]-45
theta2=data[:,1]-45
dist=getDistance(data[:,2])
```

```python
tf_2 = np.array([np.zeros(dist.shape), dist, np.zeros(dist.shape)]).T
theta_mat = np.array([theta1, theta2]).T * (np.pi/180)
theta_mat
```

```python
r = Rotation.from_euler('xz', theta_mat)
r.apply(tf_2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# ax.scatter(tf_2[:,0], tf_2[:,1], tf_2[:,2])
ax.scatter(theta1, theta2, dist)
plt.xlabel("yoink")
plt.ylabel("yeet")
```
